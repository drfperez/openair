```python
# 1️⃣ Pujar fitxer
from google.colab import files
uploaded = files.upload()

import pandas as pd

# 2️⃣ Llegir CSV
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename, sep=None, engine='python')

# 3️⃣ Convertir data automàticament (accepta .000 o ,000)
df['data_lectura'] = pd.to_datetime(
    df['data_lectura'],
    format='ISO8601',   # <- clau per evitar errors
    errors='coerce'
)

# Eliminar files amb data no vàlida (si n'hi ha)
df = df.dropna(subset=['data_lectura'])

# 4️⃣ Mantenir només hores exactes
df = df[
    (df['data_lectura'].dt.minute == 0) &
    (df['data_lectura'].dt.second == 0)
]

# 5️⃣ Crear columna date en format correcte
df['date'] = df['data_lectura'].dt.strftime('%Y-%m-%d %H:%M:%S')

# 6️⃣ Pivotar
df_pivot = df.pivot(
    index='date',
    columns='codi_variable',
    values='valor_lectura'
)

# 7️⃣ Assegurar ordre de columnes
columnes = [30,31,32,33,34,35,36]
df_pivot = df_pivot.reindex(columns=columnes)

# 8️⃣ Reset index
df_pivot = df_pivot.reset_index()

# 9️⃣ Ordenar per data
df_pivot = df_pivot.sort_values('date')

# 🔟 Guardar CSV final
output_filename = 'output_formatat.csv'
df_pivot.to_csv(output_filename, index=False)

files.download(output_filename)

df_pivot.head()
```
