```python

##############################
# 31. COMPROVACIÓ DE VARIABLES METEOROLÒGIQUES
##############################

# Llista de camps disponibles (contaminants + Meteocat)
names(ciutatwide)

##############################
# 32. ROSA DE VENT (METEOCAT)
##############################

windRose(
  ciutatwide,
  ws = "ws",
  wd = "wd",
  main = "Rosa de vents Meteocat (1991-2025)"
)

##############################
# 33. ROSA DE CONTAMINACIÓ NO2 + VENT
##############################

pollutionRose(
  ciutatwide,
  pollutant = "no2",
  ws = "ws",
  wd = "wd",
  main = "NO2 segons vent (XPVCA + Meteocat)"
)

##############################
# 34. ROSA DE CONTAMINACIÓ PM10 + VENT
##############################

pollutionRose(
  ciutatwide,
  pollutant = "pm10",
  ws = "ws",
  wd = "wd",
  main = "PM10 segons vent (XPVCA + Meteocat)"
)

##############################
# 35. EFECTE DE LA VELOCITAT DEL VENT
##############################

scatterPlot(
  ciutatwide,
  x = "ws",
  y = "pm10",
  smooth = TRUE,
  main = "Relació velocitat del vent – PM10"
)

##############################
# 36. TEMPERATURA I OZÓ (O3)
##############################

scatterPlot(
  ciutatwide,
  x = "temp",
  y = "o3",
  smooth = TRUE,
  main = "Relació temperatura – O3"
)

##############################
# 37. HUMITAT RELATIVA I PARTÍCULES
##############################

scatterPlot(
  ciutatwide,
  x = "rh",
  y = "pm2.5",
  smooth = TRUE,
  main = "Humitat relativa vs PM2.5"
)

##############################
# 38. PRESSIÓ ATMOSFÈRICA I NO2
##############################

scatterPlot(
  ciutatwide,
  x = "pressure",
  y = "no2",
  smooth = TRUE,
  main = "Pressió atmosfèrica i NO2"
)

##############################
# 39. EFECTE DE LA PLUJA EN PM10
##############################

# Classificació dies amb i sense pluja
ciutatwide$pluja <- ifelse(ciutatwide$rain > 0, "Pluja", "Sense pluja")

boxPlot(
  ciutatwide,
  pollutant = "pm10",
  group = "pluja",
  main = "Efecte de la pluja en el PM10"
)

##############################
# 40. MODEL CONCEPTUAL METEO + CONTAMINACIÓ
##############################

# Correlació entre variables meteorològiques i contaminants
corPlot(
  ciutatwide,
  pollutant = c("no2","pm10","o3","ws","temp","rh","pressure"),
  main = "Interacció Meteocat – XPVCA"
)
```
