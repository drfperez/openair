```python
# Instalar tidyverse si no lo tienes
# install.packages("tidyverse")

library(tidyverse)

# Carpeta donde están tus CSV
folder <- "sivic_data/"

# Leer todos los CSV de la carpeta y unirlos en un solo dataframe
all_data <- list.files(path = folder, pattern = "\\.csv$", full.names = TRUE) %>%
  map_dfr(read_csv)

# Ver primeras filas
head(all_data)

# Guardar CSV unificado
write_csv(all_data, "sivic_data_unido.csv")
```
