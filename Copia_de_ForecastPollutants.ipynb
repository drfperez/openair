```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import numpy as np

# Carrega les dades des del fitxer CSV
file_path = "martorelltot.csv"
df = pd.read_csv(file_path)

# Elimina files amb valors buits a la columna objectiu "no2"
df = df.dropna(subset=['no2'])

# Assegura't que hi ha prou dades per dividir
if len(df) > 0:
    # Divideix les dades en conjunts d'entrenament i prova
    X = df[['no', 'h2s', 'pm10', 'co', 'so2', 'hcnm', 'hct', 'wd', 'ws']]
    y = df['no2']

    # Utilitza train_test_split amb un test_size més petit
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    # Crea un pipeline amb l'imputador i el model de regressió lineal
    pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('regressor', LinearRegression())
    ])

    # Entrena el model i imputa les dades de manera simultània
    pipeline.fit(X_train, y_train)

    # Prediu les dades de prova
    y_pred = pipeline.predict(X_test)

    # Avalua el rendiment del model
    mse = mean_squared_error(y_test, y_pred)
    print(f'Mean Squared Error: {mse}')

    # Gràfic de comparació entre les prediccions i els valors reals amb línia de regressió
    plt.scatter(y_test, y_pred)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', linewidth=2)
    plt.xlabel('Valors Reals')
    plt.ylabel('Prediccions')
    plt.title('Comparació entre Valors Reals i Prediccions')

    # Coeficient de correlació
    correlation_coef = np.corrcoef(y_test, y_pred)[0, 1]
    plt.text(min(y_test), max(y_pred), f'Correlació: {correlation_coef:.2f}', fontsize=10, color='green')

    plt.show()

else:
    print("No hi ha prou dades per dividir.")
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Carrega les dades des del fitxer CSV
file_path = "martorelltot.csv"
df = pd.read_csv(file_path)

# Elimina files amb valors buits a la columna objectiu "no2"
df = df.dropna(subset=['no2'])

# Assegura't que hi ha prou dades per dividir
if len(df) > 0:
    # Divideix les dades en conjunts d'entrenament i prova
    X = df[['no', 'h2s', 'pm10', 'co', 'so2', 'hcnm', 'hct', 'wd', 'ws']]
    y = df['no2']

    # Utilitza train_test_split amb un test_size més petit
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    # Crea un pipeline amb l'imputador i l'estandarditzador
    pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    # Imputa i estandarditza les dades d'entrenament
    X_train_preprocessed = pipeline.fit_transform(X_train)

    # Crea una xarxa neuronal
    model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=[X_train_preprocessed.shape[1]]),
        layers.Dense(32, activation='relu'),
        layers.Dense(1)  # Capa de sortida sense funció d'activació per a regressió
    ])

    # Compila el model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Entrena el model
    model.fit(X_train_preprocessed, y_train, epochs=50, batch_size=32, validation_split=0.1, verbose=0)

    # Imputa i estandarditza les dades de prova
    X_test_preprocessed = pipeline.transform(X_test)

    # Prediu les dades de prova
    y_pred = model.predict(X_test_preprocessed).flatten()

    # Avalua el rendiment del model
    mse = mean_squared_error(y_test, y_pred)
    print(f'Mean Squared Error: {mse}')

    # Gràfic de comparació entre les prediccions i els valors reals amb línia de regressió
    plt.scatter(y_test, y_pred)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', linewidth=2)
    plt.xlabel('Valors Reals')
    plt.ylabel('Prediccions')
    plt.title('Comparació entre Valors Reals i Prediccions')

    # Coeficient de correlació
    correlation_coef = np.corrcoef(y_test, y_pred)[0, 1]
    plt.text(min(y_test), max(y_pred), f'Correlació: {correlation_coef:.2f}', fontsize=10, color='green')

    plt.show()

else:
    print("No hi ha prou dades per dividir.")
```

```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import r2_score
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# Carrega les dades des del fitxer CSV
file_path = "martorelltot.csv"
df = pd.read_csv(file_path)

# Elimina files amb valors buits a la columna objectiu "no2"
df = df.dropna(subset=['no2'])

# Assegura't que hi ha prou dades per dividir
if len(df) > 0:
    # Divideix les dades en conjunts d'entrenament i prova
    X = df[['no', 'h2s', 'pm10', 'co', 'so2', 'hcnm', 'hct', 'wd', 'ws']]
    y = df['no2']

    # Utilitza train_test_split amb un test_size més petit
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    # Imputa i estandarditza les dades
    imputer = SimpleImputer(strategy='mean')
    X_train = imputer.fit_transform(X_train)
    X_test = imputer.transform(X_test)

    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Transforma les dades per a la entrada de LSTM
    X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
    X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

    # Crea el model LSTM
    model = keras.Sequential([
        layers.LSTM(50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
        layers.Dense(1)
    ])

    # Compila el model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Entrena el model
    model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, verbose=0)

    # Avalua el rendiment del model
    mse = model.evaluate(X_test, y_test)
    print(f'Mean Squared Error: {mse}')

    # Prediu les dades de prova
    y_pred = model.predict(X_test).flatten()

    # Gràfic de comparació entre les prediccions i els valors reals amb línia de regressió
    plt.scatter(y_test, y_pred)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', linewidth=2)
    plt.xlabel('Valors Reals')
    plt.ylabel('Prediccions')
    plt.title('Comparació entre Valors Reals i Prediccions')

    # Coeficient de correlació
    correlation_coef = np.corrcoef(y_test, y_pred)[0, 1]
    plt.text(min(y_test), max(y_pred), f'Correlació: {correlation_coef:.2f}', fontsize=10, color='green')

    plt.show()

else:
    print("No hi ha prou dades per dividir.")
```

```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Carrega les dades des del fitxer CSV
file_path = "martorelltot.csv"
df = pd.read_csv(file_path)

# Elimina files amb valors buits a la columna objectiu "no2"
df = df.dropna(subset=['no2'])

# Filtra les dades per als darrers 3 mesos
end_date = pd.to_datetime(df['date']).max()  # Última data disponible
start_date = end_date - pd.DateOffset(months=3)  # Darrers 3 mesos

df_last_3_months = df[(pd.to_datetime(df['date']) >= start_date) & (pd.to_datetime(df['date']) <= end_date)]

# Assegura't que hi ha prou dades per dividir
if len(df_last_3_months) > 0:
    # Divideix les dades en conjunts d'entrenament i prova
    X = df_last_3_months[['no', 'h2s', 'pm10', 'co', 'so2', 'hcnm', 'hct', 'wd', 'ws']]
    y = df_last_3_months['no2']

    # Utilitza train_test_split amb un test_size més petit
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, shuffle=False)

    # Imputa les dades
    imputer = SimpleImputer(strategy='mean')
    X_train = imputer.fit_transform(X_train)
    X_test = imputer.transform(X_test)

    # Crea i entrena el model GBR
    model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
    model.fit(X_train, y_train)

    # Prediu les dades de prova
    y_pred = model.predict(X_test)

    # Avalua el rendiment del model
    mse = mean_squared_error(y_test, y_pred)
    print(f'Mean Squared Error: {mse}')

    # Combina les dates d'entrenament i prova per a la representació temporal
    all_dates = pd.to_datetime(df_last_3_months['date'])
    all_dates_train = all_dates.iloc[:len(X_train)]
    all_dates_test = all_dates.iloc[len(X_train):]

    # Combina els valors reals i predits
    all_values_train = y_train.values
    all_values_test = y_pred

    # Gràfic temporal
    plt.figure(figsize=(12, 6))
    plt.plot(all_dates_train, all_values_train, label='Valors Reals (Entrenament)', marker='o', linestyle='-', color='blue')
    plt.plot(all_dates_test, all_values_test, label='Valors Predits (Prova)', marker='o', linestyle='-', color='orange')
    plt.xlabel('Data')
    plt.ylabel('Valors')
    plt.title('Representació Temporal de Valors Reals i Predits amb GBR (Darrers 3 Mesos)')
    plt.legend()
    plt.show()

else:
    print("No hi ha prou dades per dividir.")
```

```python

import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Carrega les dades des del fitxer CSV
file_path = "martorelltot.csv"
df = pd.read_csv(file_path)

# Elimina files amb valors buits a la columna objectiu "no2"
df = df.dropna(subset=['no2'])

# Filtra les dades per als darrers 3 mesos
end_date = pd.to_datetime(df['date']).max()  # Última data disponible
start_date = end_date - pd.DateOffset(months=3)  # Darrers 3 mesos

df_train = df[pd.to_datetime(df['date']) < start_date]
df_last_3_months = df[(pd.to_datetime(df['date']) >= start_date) & (pd.to_datetime(df['date']) <= end_date)]

# Assegura't que hi ha prou dades per entrenar
if len(df_train) > 0:
    # Divideix les dades en conjunts d'entrenament
    X_train = df_train[['no', 'h2s', 'pm10', 'co', 'so2', 'hcnm', 'hct', 'wd', 'ws']]
    y_train = df_train['no2']

    # Imputa les dades d'entrenament
    imputer = SimpleImputer(strategy='mean')
    X_train = imputer.fit_transform(X_train)

    # Crea i entrena el model GBR amb totes les dades excepte els darrers 3 mesos
    model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
    model.fit(X_train, y_train)

    # Prediu les dades dels darrers 3 mesos
    X_last_3_months = df_last_3_months[['no', 'h2s', 'pm10', 'co', 'so2', 'hcnm', 'hct', 'wd', 'ws']]
    X_last_3_months = imputer.transform(X_last_3_months)
    y_pred_last_3_months = model.predict(X_last_3_months)

    # Combina les dates reals i previstes dels darrers 3 mesos per a la representació temporal
    all_dates_last_3_months = pd.to_datetime(df_last_3_months['date'])
    all_values_real_last_3_months = df_last_3_months['no2'].values

    # Gràfic temporal amb valors reals i prevists dels darrers 3 mesos
    plt.figure(figsize=(12, 6))
    plt.plot(all_dates_last_3_months, all_values_real_last_3_months, label='Valors Reals', marker='o', linestyle='-', color='blue')
    plt.plot(all_dates_last_3_months, y_pred_last_3_months, label='Valors Predits', marker='o', linestyle='-', color='orange')
    plt.xlabel('Data')
    plt.ylabel('Valors')
    plt.title('Representació Temporal de Valors Reals i Predits amb GBR (Darrers 3 Mesos)')
    plt.legend()
    plt.show()

else:
    print("No hi ha prou dades per entrenar.")
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Carrega les dades des del fitxer CSV
file_path = "martorelltot.csv"
df = pd.read_csv(file_path)

# Assegura't que la columna de dates sigui de tipus datetime
df['date'] = pd.to_datetime(df['date'])

# Filtra les dades per a aquelles amb mesures de NO2
df_no2 = df.dropna(subset=['no2'])

# Calcula la mitjana anual de NO2 per cada any
annual_mean_no2 = df_no2.groupby(df_no2['date'].dt.year)['no2'].mean()

# Calcula la diària màxima de NO2 per cada dia
daily_max_no2 = df_no2.groupby(df_no2['date'])['no2'].max()

# Defineix els llindars de la normativa europea per NO2
annual_limit = 40  # µg/m³
daily_limit = 200  # µg/m³

# Comprova si es compleixen els llindars
annual_compliance = all(annual_mean_no2 <= annual_limit)
daily_compliance = all(daily_max_no2 <= daily_limit)

# Imprimeix si es compleixen els llindars
print(f"Compliment de la normativa europea de NO2:")
print(f"- Mitjana Anual (Límit: {annual_limit} µg/m³): {'Sí' if annual_compliance else 'No'}")
print(f"- Diària Màxima (Límit: {daily_limit} µg/m³): {'Sí' if daily_compliance else 'No'}")

# Gràfic de la mitjana anual amb límit
plt.figure(figsize=(10, 5))
plt.plot(annual_mean_no2.index, annual_mean_no2, label='Mitjana Anual NO2', marker='o', linestyle='-', color='blue')
plt.axhline(y=annual_limit, color='red', linestyle='--', label='Límit Mitjana Anual NO2')
plt.xlabel('Any')
plt.ylabel('Concentració de NO2 (µg/m³)')
plt.title('Mitjana Anual de NO2 amb Límit Europeu')
plt.legend()
plt.show()

# Gràfic de la diària màxima amb límit
plt.figure(figsize=(10, 5))
plt.plot(daily_max_no2.index, daily_max_no2, label='Diària Màxima NO2', marker='o', linestyle='-', color='green')
plt.axhline(y=daily_limit, color='orange', linestyle='--', label='Límit Diària Màxima NO2')
plt.xlabel('Data')
plt.ylabel('Concentració de NO2 (µg/m³)')
plt.title('Diària Màxima de NO2 amb Límit Europeu')
plt.legend()
plt.show()
```
