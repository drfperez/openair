```python

# Importar biblioteques
import pandas as pd
import numpy as np
from google.colab import files

# 📌 1. Pujar el fitxer CSV
uploaded = files.upload()
file_name = next(iter(uploaded))  # Obtenir el nom del fitxer pujat
df = pd.read_csv(file_name)

# 🔍 2. Mostrar les primeres files abans de la conversió
print("📌 Primeres files abans de la conversió:")
print(df.head())

# 🔤 3. Convertir els noms de les columnes a minúscules
df.columns = df.columns.str.lower()

# 🗓️ 4. Convertir la columna "data" a format ISO (yyyy-mm-dd hh:mm:ss)
df['data'] = pd.to_datetime(df['data']).dt.strftime('%Y-%m-%d %H:%M:%S')

# 🎯 5. Seleccionar només les columnes rellevants per al pivotatge
columnes_utilitzades = ['data', 'contaminant'] + [f'h{i:02d}' for i in range(1, 25)]
df = df[columnes_utilitzades]

# 🔄 6. Pivotar: transformar hores en files
df_pivotat = df.melt(id_vars=['data', 'contaminant'],
                     var_name='hora',
                     value_name='valor')

# 🕒 7. Convertir la columna "hora" en format correcte (ex: 'h01' → '01:00:00')
df_pivotat['hora'] = df_pivotat['hora'].str.extract('(\d+)').astype(int)
df_pivotat['hora'] = df_pivotat['hora'].astype(str).str.zfill(2) + ':00:00'

# 📅 8. Crear la nova columna "date" combinant "data" + "hora"
df_pivotat['date'] = df_pivotat['data'].str[:10] + ' ' + df_pivotat['hora']

# 🧹 9. Eliminar les columnes antigues de data i hora
df_pivotat = df_pivotat[['date', 'contaminant', 'valor']]

# 🔄 10. Pivotar perquè cada contaminant sigui una columna
df_final = df_pivotat.pivot(index='date', columns='contaminant', values='valor')

# 🔤 11. Convertir els noms de columna a minúscules i ordenar-los
df_final.columns = df_final.columns.str.lower()
df_final = df_final.reset_index()  # Recuperar "date" com a columna

# 🚀 12. Substituir valors buits per 'NA'
df_final = df_final.fillna('NA')

# 🔍 13. Mostrar les primeres files després de la conversió
print("\n✅ Primeres files després de la conversió:")
print(df_final.head())

# 💾 14. Guardar el fitxer com a CSV
output_file = '/content/dadeshoraries.csv'
df_final.to_csv(output_file, index=False)

# 📥 15. Descarregar automàticament el fitxer CSV
files.download(output_file)
```

```python

# Importar les biblioteques necessàries
import pandas as pd
import numpy as np
from google.colab import files

# 📌 1. Pujar el fitxer CSV
uploaded = files.upload()
file_name = next(iter(uploaded))  # Obtenir el nom del fitxer pujat
df = pd.read_csv(file_name)

# 🔍 2. Mostrar les primeres files abans de la conversió
print("📌 Primeres files abans de la conversió:")
print(df.head())

# 🔤 3. Convertir els noms de les columnes a minúscules
df.columns = df.columns.str.lower()

# 🗓️ 4. Convertir la columna "data" a format ISO (yyyy-mm-dd hh:mm:ss)
df['data'] = pd.to_datetime(df['data']).dt.strftime('%Y-%m-%d %H:%M:%S')

# 🚀 5. Substituir valors buits: 'NA' per columnes de text, np.nan per columnes numèriques
for col in df.columns:
    if df[col].dtype == 'object':  # Si és text
        df[col].fillna('NA', inplace=True)
    else:  # Si és numèric
        df[col].fillna(np.nan, inplace=True)

# 🎯 6. Seleccionar només les columnes rellevants per al pivotatge
columnes_utilitzades = ['data', 'contaminant'] + [f'h{i:02d}' for i in range(1, 25)]
df = df[columnes_utilitzades]

# 🔄 7. Pivotar: transformar hores en files
df_pivotat = df.melt(id_vars=['data', 'contaminant'],
                     var_name='hora',
                     value_name='valor')

# 🕒 8. Convertir la columna "hora" en format correcte (ex: 'h01' → '01:00:00')
df_pivotat['hora'] = df_pivotat['hora'].str.extract('(\d+)').astype(int)
df_pivotat['hora'] = df_pivotat['hora'].astype(str).str.zfill(2) + ':00:00'

# 📅 9. Crear la nova columna "date" combinant "data" + "hora"
df_pivotat['date'] = df_pivotat['data'].str[:10] + ' ' + df_pivotat['hora']

# 🧹 10. Eliminar les columnes antigues de data i hora
df_pivotat = df_pivotat[['date', 'contaminant', 'valor']]

# 🔄 11. Pivotar perquè cada contaminant sigui una columna
df_final = df_pivotat.pivot(index='date', columns='contaminant', values='valor')

# 🔤 12. Convertir els noms de columna a minúscules i ordenar-los
df_final.columns = df_final.columns.str.lower()
df_final = df_final.reset_index()  # Recuperar "date" com a columna

# 🔍 13. Mostrar les primeres files després de la conversió
print("\n✅ Primeres files després de la conversió:")
print(df_final.head())

# 💾 14. Guardar el fitxer com a CSV
output_file = '/content/dadeshoraries.csv'
df_final.to_csv(output_file, index=False)

# 📥 15. Descarregar automàticament el fitxer CSV
files.download(output_file)
```
