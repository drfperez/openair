```python

# Importar biblioteques
import pandas as pd
import numpy as np
from google.colab import files

# üìå 1. Pujar el fitxer CSV
uploaded = files.upload()
file_name = next(iter(uploaded))  # Obtenir el nom del fitxer pujat
df = pd.read_csv(file_name)

# üîç 2. Mostrar les primeres files abans de la conversi√≥
print("üìå Primeres files abans de la conversi√≥:")
print(df.head())

# üî§ 3. Convertir els noms de les columnes a min√∫scules
df.columns = df.columns.str.lower()

# üóìÔ∏è 4. Convertir la columna "data" a format ISO (yyyy-mm-dd hh:mm:ss)
df['data'] = pd.to_datetime(df['data']).dt.strftime('%Y-%m-%d %H:%M:%S')

# üéØ 5. Seleccionar nom√©s les columnes rellevants per al pivotatge
columnes_utilitzades = ['data', 'contaminant'] + [f'h{i:02d}' for i in range(1, 25)]
df = df[columnes_utilitzades]

# üîÑ 6. Pivotar: transformar hores en files
df_pivotat = df.melt(id_vars=['data', 'contaminant'],
                     var_name='hora',
                     value_name='valor')

# üïí 7. Convertir la columna "hora" en format correcte (ex: 'h01' ‚Üí '01:00:00')
df_pivotat['hora'] = df_pivotat['hora'].str.extract('(\d+)').astype(int)
df_pivotat['hora'] = df_pivotat['hora'].astype(str).str.zfill(2) + ':00:00'

# üìÖ 8. Crear la nova columna "date" combinant "data" + "hora"
df_pivotat['date'] = df_pivotat['data'].str[:10] + ' ' + df_pivotat['hora']

# üßπ 9. Eliminar les columnes antigues de data i hora
df_pivotat = df_pivotat[['date', 'contaminant', 'valor']]

# üîÑ 10. Pivotar perqu√® cada contaminant sigui una columna
df_final = df_pivotat.pivot(index='date', columns='contaminant', values='valor')

# üî§ 11. Convertir els noms de columna a min√∫scules i ordenar-los
df_final.columns = df_final.columns.str.lower()
df_final = df_final.reset_index()  # Recuperar "date" com a columna

# üöÄ 12. Substituir valors buits per 'NA'
df_final = df_final.fillna('NA')

# üîç 13. Mostrar les primeres files despr√©s de la conversi√≥
print("\n‚úÖ Primeres files despr√©s de la conversi√≥:")
print(df_final.head())

# üíæ 14. Guardar el fitxer com a CSV
output_file = '/content/dadeshoraries.csv'
df_final.to_csv(output_file, index=False)

# üì• 15. Descarregar autom√†ticament el fitxer CSV
files.download(output_file)
```

```python

# Importar les biblioteques necess√†ries
import pandas as pd
import numpy as np
from google.colab import files

# üìå 1. Pujar el fitxer CSV
uploaded = files.upload()
file_name = next(iter(uploaded))  # Obtenir el nom del fitxer pujat
df = pd.read_csv(file_name)

# üîç 2. Mostrar les primeres files abans de la conversi√≥
print("üìå Primeres files abans de la conversi√≥:")
print(df.head())

# üî§ 3. Convertir els noms de les columnes a min√∫scules
df.columns = df.columns.str.lower()

# üóìÔ∏è 4. Convertir la columna "data" a format ISO (yyyy-mm-dd hh:mm:ss)
df['data'] = pd.to_datetime(df['data']).dt.strftime('%Y-%m-%d %H:%M:%S')

# üöÄ 5. Substituir valors buits: 'NA' per columnes de text, np.nan per columnes num√®riques
for col in df.columns:
    if df[col].dtype == 'object':  # Si √©s text
        df[col].fillna('NA', inplace=True)
    else:  # Si √©s num√®ric
        df[col].fillna(np.nan, inplace=True)

# üéØ 6. Seleccionar nom√©s les columnes rellevants per al pivotatge
columnes_utilitzades = ['data', 'contaminant'] + [f'h{i:02d}' for i in range(1, 25)]
df = df[columnes_utilitzades]

# üîÑ 7. Pivotar: transformar hores en files
df_pivotat = df.melt(id_vars=['data', 'contaminant'],
                     var_name='hora',
                     value_name='valor')

# üïí 8. Convertir la columna "hora" en format correcte (ex: 'h01' ‚Üí '01:00:00')
df_pivotat['hora'] = df_pivotat['hora'].str.extract('(\d+)').astype(int)
df_pivotat['hora'] = df_pivotat['hora'].astype(str).str.zfill(2) + ':00:00'

# üìÖ 9. Crear la nova columna "date" combinant "data" + "hora"
df_pivotat['date'] = df_pivotat['data'].str[:10] + ' ' + df_pivotat['hora']

# üßπ 10. Eliminar les columnes antigues de data i hora
df_pivotat = df_pivotat[['date', 'contaminant', 'valor']]

# üîÑ 11. Pivotar perqu√® cada contaminant sigui una columna
df_final = df_pivotat.pivot(index='date', columns='contaminant', values='valor')

# üî§ 12. Convertir els noms de columna a min√∫scules i ordenar-los
df_final.columns = df_final.columns.str.lower()
df_final = df_final.reset_index()  # Recuperar "date" com a columna

# üîç 13. Mostrar les primeres files despr√©s de la conversi√≥
print("\n‚úÖ Primeres files despr√©s de la conversi√≥:")
print(df_final.head())

# üíæ 14. Guardar el fitxer com a CSV
output_file = '/content/dadeshoraries.csv'
df_final.to_csv(output_file, index=False)

# üì• 15. Descarregar autom√†ticament el fitxer CSV
files.download(output_file)
```
