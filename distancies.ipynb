```python

# Instal·lem llibreries necessàries
!pip install pandas geopy numpy --quiet

import pandas as pd
import numpy as np
from geopy.distance import geodesic
from google.colab import files

# ----------------------
# Pas 1: Crear DataFrame amb les dades
# ----------------------

# Llista neta de ciutats amb CODI, LLOC, LAT, LONG
data = [
    ["AN", "Av. Lluis Companys[AN", 41.39004, 2.18091],
    ["D5", "Barcelona - Observatori Fabra", 41.41864, 2.12379],
    ["X2", "Barcelona - Zoo", 41.38943, 2.18847],
    ["X4", "Barcelona - el Raval", 41.38390, 2.16775],
    ["X8", "Barcelona - Zona Universitària", 41.37919, 2.10540],
    ["Y9", "Port de Barcelona - ZAL Prat (ca l'Arana)", 41.31723, 2.13091],
    ["YQ", "Port de Barcelona - ZAL Prat", 41.31928, 2.13720],
    ["Y7", "Port de Barcelona - Bocana Sud", 41.31725, 2.16537],
    ["08019044", "Barcelona (Gràcia - Sant Gervasi)", 41.398724, 2.1533988],
    ["08019004", "Barcelona Poblenou", 41.40388, 2.204501],
    ["08019058", "Barcelona (observatori-Fabra)", 41.41843, 2.1238973],
    ["08019043", "Eixample", 41.385315, 2.1537998],
    ["08019054", "Parc Vall Hebron", 41.42611, 2.1480017],
    ["08019057", "Barcelona (Palau Reial)", np.nan, np.nan],  # Lat/Long no disponible
    ["08019042", "Barcelona (Sants)", 41.37878, 2.133099]
]

# Convertim a DataFrame
df = pd.DataFrame(data, columns=["CODI", "LLOC", "LAT", "LONG"])

# Eliminem files sense lat/long
df = df.dropna(subset=["LAT", "LONG"]).reset_index(drop=True)

# ----------------------
# Pas 2: Crear taula de distàncies optimitzada
# ----------------------
cities = df['LLOC'].tolist()
coords = list(zip(df['LAT'], df['LONG']))
n = len(cities)

# Creamos un array buit
distances = np.zeros((n, n))

# Omplim la taula simètrica de distàncies
for i in range(n):
    for j in range(i, n):
        dist = geodesic(coords[i], coords[j]).kilometers
        distances[i,j] = dist
        distances[j,i] = dist

# Convertim a DataFrame amb ciutats com a index i columnes
distances_df = pd.DataFrame(distances, index=cities, columns=cities)

# Mostrar tota la taula sense truncar
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
print(distances_df)

# ----------------------
# Pas 3: Guardar CSV i permetre descàrrega
# ----------------------
output_filename = "distancies_ciutats_barcelona.csv"
distances_df.to_csv(output_filename)
files.download(output_filename)
```

```python

# Instal·lem llibreries necessàries
!pip install pandas geopy numpy --quiet

import pandas as pd
import numpy as np
from geopy.distance import geodesic
from google.colab import files

# ----------------------
# Pas 1: Pujar CSV
# ----------------------
uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Comprovem columnes
assert {'ciutat','lat','long'}.issubset(df.columns), "El CSV ha de tenir columnes: ciutat, lat, long"

cities = df['ciutat'].tolist()
coords = list(zip(df['lat'], df['long']))
n = len(cities)

# ----------------------
# Pas 2: Crear taula de distàncies optimitzada
# ----------------------
# Creamos un array buit
distances = np.zeros((n, n))

# Omplim només la meitat superior (matriu simètrica)
for i in range(n):
    for j in range(i, n):
        dist = geodesic(coords[i], coords[j]).kilometers
        distances[i,j] = dist
        distances[j,i] = dist  # simètrica

# Convertim a DataFrame
distances_df = pd.DataFrame(distances, index=cities, columns=cities)

# Mostrar tota la taula sense truncar
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
print(distances_df)

# ----------------------
# Pas 3: Guardar CSV i permetre descàrrega
# ----------------------
output_filename = "distancies_ciutats_optimitzat.csv"
distances_df.to_csv(output_filename)
files.download(output_filename)
```
